```python
import math
import pandas as pd
import numpy as np
from datetime import datetime
from tqdm.notebook import tqdm
tqdm.pandas()
```

```python
customers_df = pd.read_csv("../data/olist_customers_dataset.csv")
geoloc_df = pd.read_csv("../data/olist_geolocation_dataset.csv")
order_items_df = pd.read_csv("../data/olist_order_items_dataset.csv")
order_payments_df = pd.read_csv("../data/olist_order_payments_dataset.csv")
order_reviews_df = pd.read_csv("../data/olist_order_reviews_dataset.csv")
order_df = pd.read_csv("../data/olist_orders_dataset.csv")
products_df = pd.read_csv("../data/olist_products_dataset.csv")
sellers_df = pd.read_csv("../data/olist_sellers_dataset.csv")
product_category_name_translation_df = pd.read_csv("../data/product_category_name_translation.csv")
```

```python
orders_products_reviews_df = order_df.merge(
    order_reviews_df.merge(
            order_items_df.merge(products_df,
                                 on="product_id", 
                                 how='left'),
            on="order_id",
            how="left"),
    on="order_id",
    how="left"
)
```

```python
orders_products_reviews_df.reset_index(drop=True, inplace=True)
```

```python
orders_products_reviews_df.to_csv('../data/orders_products_reviews_dataset.csv', index=False)
```

```python
# orders_products_reviews_df.isna().sum()
orders_products_reviews_df = orders_products_reviews_df.select_dtypes(include=[np.number])
```

```python
# orders_products_reviews_df

orders_products_reviews_df["product_weight_g"] = orders_products_reviews_df["product_weight_g"].fillna(orders_products_reviews_df["product_weight_g"].median())
orders_products_reviews_df["product_length_cm"] = orders_products_reviews_df["product_length_cm"].fillna(orders_products_reviews_df["product_length_cm"].median())
orders_products_reviews_df["product_height_cm"] = orders_products_reviews_df["product_height_cm"].fillna(orders_products_reviews_df["product_height_cm"].median())
orders_products_reviews_df["product_width_cm"] = orders_products_reviews_df["product_width_cm"].fillna(orders_products_reviews_df["product_width_cm"].median())
# orders_products_reviews_df["review_comment_message"].fillna("No review", inplace=True)
```
